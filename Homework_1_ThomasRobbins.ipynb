#### Isolate Idaho data
Next, to isolate the Idaho data, I initialized all the values in the ‘state’ column that had a value of ID as ‘IData’. Then, I took all those values in IData and initialized the values in the U_ppm column with the heading ‘Id_U’. This allowed me to run statistical analysis on the uranium concentrations from all the samples collected in Idaho. 

```python
IData = data.loc[data['State'] == 'ID']
Id_U = IData['U_ppm'].values
Id_U.max()
```

#### Scatter Plot
Next, I attempted to see where the highest concentration was based on the latitude and longitude provided from the csv file. The scatter plot shows the concentration of uranium based on color shading in the graph. Lighter coloring indicates a lower concertation, and darker coloring is high concentration. Also, size of the plotted circles indicates concentration levels. 
From the provided Lat and Long, the sample locations surround the Sawtooth national forest in Idaho. 

```python
Id_Lat = IData['Lat_NAD27'].values
Id_Long = IData['Long_NAD27'].values
plt.scatter(Id_Long, Id_Lat, Id_U, alpha = 0.5)

plt.title('Uranium ppm in Idaho')
plt.legend('loc=2')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
# plt.colorbar(mappable, **kwargs)
plt.show
```

### Scatter plot with color
This is an attempt to add color to the scatter plot. The large red mark on the graph indicates the point with the highest concentration.

```python
color_by = Id_U
label = 'Uranium ppm'
max_color_by = max(color_by)
min_color_by = min(color_by)

fig, ax = plt.subplots()
s = ax.scatter(Id_Long, Id_Lat, Id_U,
               c=color_by,
               marker='o',               
               alpha = 0.5,
               cmap = plt.cm.coolwarm,      
               vmin = min_color_by,          
               vmax = max_color_by)         

cbar = plt.colorbar(mappable = s, ax = ax)   
cbar.set_label(label)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Uranimum in Idaho')
plt.show()
```
